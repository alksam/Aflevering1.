##  Imputing Missing Data


```python
import pandas as pd
```

```python
dataset1 = 'https://raw.githubusercontent.com/TrainingByPackt/Data-Science-with-Python/master/Chapter01/Data/Banking_Marketing.csv'
```

```python
df = pd.read_csv(dataset, header=0)
```

```python
# Vis de første par rækker af DataFramen for at sikre, at det er indlæst korrekt
df.head()
```

```python
# Beregn gennemsnittet af age-kolonnen
mean_age = df['age'].mean()
```

```python
print(mean_age)
```

```python
# Imputer manglende værdier i age-kolonnen ved at tildele direkte
df['age'] = df['age'].fillna(mean_age)
```

```python
print(mean_age)
```

```python
# Beregn medianen af duration-kolonnen
median_duration = df['duration'].median()
```

```python
print(median_duration)
```

```python
df['duration'] = df['duration'].fillna(median_duration)
```

```python
print(median_duration)
```

```python
# Beregn modus af contact-kolonnen
mode_contact = df['contact'].mode()[0]
```

```python
print(mode_contact)
```

```python
# Imputer manglende værdier i contact-kolonnen ved at tildele direkte
df['contact'] = df['contact'].fillna(mode_contact)
```

```python
df.fillna({'age': mean_age, 'duration': median_duration, 'contact': mode_contact}, inplace=True)
```

```python
print(mode_contact)
```

```python
print(median_duration)
```

## Finding and Removing Outliers in Data

```python
import pandas as pd

import numpy as np

%matplotlib inline

import seaborn as sbn
```

```python
dataset = 'https://raw.githubusercontent.com/TrainingByPackt/Data-Science-with-Python/master/Chapter01/Data/german_credit_data.csv'
```

```python
df = pd.read_csv(dataset, header=0)
```

```python
# Vis de første par rækker af DataFramen for at sikre, at det er indlæst korrekt
df.head()
```

```python
sbn.boxplot(df['Age'])
```

```python
Q1 = df["Age"].quantile(0.25)
```

```python
Q3 = df["Age"].quantile(0.75)
```

```python
IQR = Q3 - Q1
```

```python
print(IQR)
```

```python
Lower_Fence = Q1 - (1.5 * IQR)

Upper_Fence = Q3 + (1.5 * IQR)

print(Lower_Fence)

print(Upper_Fence)
```

```python
df[((df["Age"] < Lower_Fence) |(df["Age"] > Upper_Fence))]
```

```python
df = df[~((df ["Age"] < Lower_Fence) |(df["Age"] > Upper_Fence))]
```
